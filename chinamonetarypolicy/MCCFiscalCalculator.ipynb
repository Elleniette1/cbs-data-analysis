```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
china_lend = pd.read_csv('../data/imf-dm-export-20250211.csv')
china_lend = china_lend.loc[china_lend['Year'].between(2010,2021),:]
china_lend = china_lend.set_index('Year')
china_lend = china_lend.rename({'Year':'year'})
china_lend.plot.line(color = 'red')
```

```python
china_grouped = []
list1 = []
list2 = []
china_grouped = pd.DataFrame(china_grouped)
for i in range(0,12,3):
    list1.append(f"{2010+i}-{2010+2+i}")
    list2.append(china_lend.loc[2010+i:2010+i+2,'net_lending'].mean())
list1[-1] = "2019-2020"
china_grouped['year'] = list1
china_grouped['mean_3_years'] = list2
sns.lineplot(data=china_grouped,x='year',y='mean_3_years', color = 'red')
plt.title("MCC's Fiscal Policy Indicator:",weight = 'bold', size = 21)
plt.xlabel('Years')
plt.xticks(color = 'grey')
plt.yticks(color = 'grey')
plt.ylabel('Net Lending/Borrowing (%)')
```
